## [Example of working PDF](https://github.com/Hpilton/Tabula-py/blob/master/test.pdf)

```python
from tabula import read_pdf
```

```python
df = read_pdf("test.pdf")
df.fillna("")
```

### Tabula-py может излвекает таблицу из документа и возвращать ее в виде pd.Dataframe, как это показано выше.
###  Или же в виде json объекта, как это показано ниже.

```python
df = read_pdf("test.pdf",output_format='json')
df.fillna("")
```

### Также для более корректного извлечения, желательно указывать область в которой находится таблица, это можно делать с помощью параметра area.

```python
df = read_pdf("test.pdf",  area = [242, 0, 5000.5, 700])
df.fillna("")
```

### Или же нарезать pdf файл вручную.
### После моих исследование, нарезка документа вручную - дает лучший результат  извлечения.

```python
from PyPDF2 import PdfFileWriter, PdfFileReader
pdf_file = PdfFileReader(open("test.pdf","rb"))
page = pdf_file.getPage(0)
top = float(page.cropBox.getUpperLeft()[1]) // 1.45
bot = page.cropBox.getLowerRight()[1] // 2
left = page.cropBox.getLowerLeft()[0]
right = page.cropBox.getLowerRight()[0]
page.cropBox.setLowerLeft((left,bot))
page.cropBox.setUpperRight((right, top))
writer = PdfFileWriter()
writer.addPage(page)
outstream = open("test_final.pdf", "wb")
writer.write(outstream)
outstream.close()
df = read_pdf('test_final.pdf')
df = df.fillna("")
```

### Библиотека иногда использует «потоки» пробелов для воссоздания структуры вашей таблицы
### Tabula имеет два метода извлечения. Он пытается угадать, какой из них подходит для документа, но иногда это неправильно.
### И можно вручную  выбрать другой («stream» и «lettice») в параметере извлечения.

```python
df = read_pdf("test.pdf", stream=True,  guess = True)
df.fillna("")
```
